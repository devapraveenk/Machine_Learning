```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
country=ins_df['country'].value_counts()
country
```

```python
plt.figure(figsize=(16,9))
plt.title("Influencers Country")

sns.countplot(x=ins_df['country'])

plt.xticks(rotation=90)
```

```python
top10_ER = ins_df. drop(["rank", "influence_score", "posts", "avg_likes", "new_post_avg_like", "total_likes", "country"],axis=1)
top10_ER.head()
```

```python
replace = {'b':'e9','m':'e6','k':'e3', '%':''}
cov_followers = top10_ER['followers'].replace(replace,regex=True).astype(float)
np=top10_ER['60_day_eng_rate'].replace(replace,regex=True).astype(float)
np
```

```python

plt.figure(figsize=(40,20))
plt.pie(x=cov_followers,labeldistance=1,labels=top10_ER['channel_info'])
plt.show()
```
